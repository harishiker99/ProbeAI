```python
from datasets import Dataset
import json
import pandas as pd
from pathlib import Path
```

```python
# Load your JSON data
with open("/content/ragas_ready_dataset.json", "r") as f:
    json1 = json.load(f)
```

```python
dataset = Dataset.from_list(json1)


```

```python

# Now ready for Ragas evaluation
print(dataset)
```

```python
import os
os.environ["OPENAI_API_KEY"] = ""
```

```python
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
llm = ChatOpenAI(model="gpt-4o-mini")
embeddings = OpenAIEmbeddings()
```

```python
from ragas import evaluate
from ragas.llms import LangchainLLMWrapper
from ragas import evaluate
from ragas.metrics import (
    answer_relevancy,
    faithfulness,
    context_precision,
    context_recall,
    NoiseSensitivity
)


# Wrap the LangChain LLM
evaluator_llm = LangchainLLMWrapper(llm)

# Run evaluation with all desired metrics
result = evaluate(
    dataset=dataset,
    metrics=[
        answer_relevancy,
        faithfulness,
        context_precision,
        context_recall,
    ],
    llm=evaluator_llm
  )

result
```
